```python
#Importing Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_excel('/content/student_performance.xlsx')
```

```python
df2 = pd.read_excel('/content/car_specifications.xlsx')
```

```python
print(df1)
```

```python
print(df2)
```

```python
# Histogram
plt.figure(figsize=(8,5))
sns.histplot(data=df1, x="Test_Score", hue="Gender", kde=True, bins=6, palette="rocket")
plt.title("Distribution of Test Scores by Gender")
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(6,5))
sns.boxplot(data=df1, x="Gender", y="Test_Score", palette="rocket")
plt.title("Boxplot of Test Scores by Gender")
plt.show()
```

```python
sns.kdeplot(data=df1, x='Test_Score', fill=True)
plt.title('KDE Plot of Test Scores')
plt.show()
```

```python
#Interpretation:
##The histogram with KDE shows that female students scores are slightly shifted towards the higher end compared to male students. Female students peak around 7080, while males cluster around 5065.
#The boxplot indicates that female students have a higher median score than male students. The spread for males is wider with lower minimum values, while females are more consistent.
#A slight right skew is observed in male scores, while female scores appear more normally distributed.
#No extreme outliers are detected, though male students have more variability in their performance.
```

```python
sns.histplot(data=df1, x='Test_Score', bins=5, kde=True, hue='Gender', palette="rocket", alpha=0.6)
plt.title('Histogram + KDE by Gender')
plt.show()
```

```python
#The histogram with KDE shows that female students scores are concentrated in the higher range (6080), while male students are more spread across low and high scores.
#The male distribution is wider, indicating greater variability in their performance compared to females.
#The female KDE curve is tall and narrow, suggesting consistency in their results, whereas the male curve is flatter, showing uneven performance levels.
#Overall, female students appear to perform more steadily, while male students show greater variation with both high and low performers.
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df2, x="Engine_Size", y="Price",
                hue="Brand", size="Horsepower", palette="rocket", sizes=(50,300))
plt.title("Relationship between Engine Size and Price of Cars")
plt.show()
```

```python
#The scatter plot shows that larger engine size generally leads to higher prices, but the relationship is not strictly linear.
#Luxury brands like BMW and Audi have disproportionately higher prices even for similar engine sizes, showing that brand value strongly affects pricing.
#Horsepower (bubble size) also influences the trend: cars with higher horsepower tend to be priced higher, even when engine size is similar.
#Smaller cars like Suzuki and Toyota remain affordable despite decent mileage, while performance cars with powerful engines dominate the higher price range.
#Thus, engine size alone does not determine price  brand positioning and horsepower are also key factors.
```
